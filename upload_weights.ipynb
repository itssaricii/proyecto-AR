```python
import threading
import time
import subprocess
import os

REPO_DIR = '/home/studio-lab-user/sagemaker-studiolab-notebooks/proyecto-AR'
WEIGHTS_DIR = os.path.join(REPO_DIR, 'weights')

def auto_push_weights(interval=60):
    while True:
        try:
            print("[AutoGit] Comprobando cambios en carpeta weights...")

            # Hacer add solo de la carpeta weights
            subprocess.run(['git', '-C', REPO_DIR, 'add', 'weights/'])

            # Comprobar si hay algo que commitear
            result = subprocess.run(['git', '-C', REPO_DIR, 'status', '--porcelain'],
                                    capture_output=True, text=True)

            if result.stdout.strip():  # Si hay cambios
                subprocess.run(['git', '-C', REPO_DIR, 'commit', '-m', 'Actualizacin automtica de pesos'])
                subprocess.run(['git', '-C', REPO_DIR, 'push'])
                print("[AutoGit] Cambios subidos correctamente.")
            else:
                print("[AutoGit] No hay cambios en weights.")
        except Exception as e:
            print(f"[AutoGit] Error: {e}")

        time.sleep(interval)

# Lanzar el hilo en segundo plano
git_thread = threading.Thread(target=auto_push_weights, args=(60,), daemon=True)
git_thread.start()
```
